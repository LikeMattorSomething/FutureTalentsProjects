```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


GraceData = pd.read_csv("/content/drive/MyDrive/Future Talents Data/Grace_All years combined 08-07-2025.csv")

GraceData["Diabetes"].fillna(value=0)

for Column_Name in GraceData.columns:

  if Column_Name != 'seriali': # Exclude 'seriali' column from plotting

    print("Feature: ",Column_Name)
    print("Datatype: ",GraceData[Column_Name].dtype)

    if GraceData[Column_Name].dtype == int or GraceData[Column_Name].dtype == float:


        if GraceData[Column_Name].nunique() < 5:

          GraceData[Column_Name].value_counts().plot(kind='pie')
          display(GraceData[Column_Name].value_counts())

          plt.show()

        else:
          GraceData.plot.hist(y=Column_Name,bins=range(int(GraceData[Column_Name].max()+1)))

          plt.show()


    elif Column_Name == "Diabetes":
        TempDiabetesData = (GraceData[Column_Name].replace("#NULL!","No Data"))
        TempDiabetesData = TempDiabetesData.astype(str)
        TempDiabetesData.value_counts().plot(kind='pie')

        plt.show()

        if GraceData[Column_Name].nunique() < 5:
          display(GraceData[Column_Name].value_counts())

    else:
        GraceData[Column_Name].value_counts().plot(kind='pie')

        plt.show()

        if GraceData[Column_Name].nunique() < 5:
          display(GraceData[Column_Name].value_counts())
    print("")
```
